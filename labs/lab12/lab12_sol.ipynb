```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2 * np.pi, 10)
```

```python
y = np.sin(x)
```

```python
print(x)
```

## Wykres liniowy

```python
plt.plot(x)
plt.show() #potrzebne w niektórych IDE
```

#### modyfikacje

```python
plt.plot(x, y)
plt.ylabel('sinus')
```

```python
plt.plot(x, y)
plt.ylabel('sinus')
plt.title("SINUS", loc = "left")
plt.show()
```

```python
plt.plot(x, y, linestyle = "--")
plt.show()
```

```python
plt.plot(x, y, color = "#371ea3")
```

Więcej: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
plt.plot(x, y, label = "sin")
plt.plot(x, -y, label = "-sin")
plt.legend()
plt.show()
```

## Figury

`plt.plot()` pozwala na narysowanie tylko jednej figury o współrzędnych (x,y)

```python
names = ['X', 'Y', 'Z']
values = [3, 9, 27]
```

```python
plt.bar(names, values)
plt.show()
```

`plt.figure()` definiuje figurę, możemy na niej umieszczać inne elementy

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) #nrow, ncol, index
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1) #nrow, ncol, index
ax.bar(names, values)
plt.show()
```

`plt.subplot()` - pozwala na rysowanie jednej lub kilku figur w tym samym oknie

```python
fig, (ax1, ax2) = plt.subplots(2, 1) #2 wiersze, 1 kolumna
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Tytuł')

ax1.bar(names, values)
ax2.bar(names, [2,2,5])

plt.show()
```

# seaborn

```python
import seaborn as sns
import pandas as pd
```

```python
# opcje wyświetlania
```

```python
sns.set(rc = {'figure.figsize': (12,7)})
sns.set_style("darkgrid")
pd.options.display.max_rows = 10
pd.set_option('display.width', 5000)
pd.set_option('display.max_columns', 60)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mini-pw/2021Z-DataVisualizationTechniques/master/labs/data/Pokemon.csv", index_col=0, sep=",")
```

```python
df.head()
```

## Wykres punktowy

```python
sns.scatterplot(data=df, x='Attack', y='Defense')
plt.show()
```

```python
sns.scatterplot(data=df, x='Attack', y='Defense', hue='Stage')
plt.show()
```

```python
sns.scatterplot(data=df, x='Attack', y='Defense', size='Stage')
plt.show()
```

```python
sns.scatterplot(data=df, x='Attack', y='Defense', hue='Stage')

plt.ylim(None, 150)
plt.xlim(-100, None)
plt.show()
```

## boxplot

```python
sns.boxplot(data=df)
plt.show()
```

```python
sns.boxplot(data=df, color='w')
plt.show()
```

```python
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
```

```python
sns.boxplot(data=stats_df)
```

```python
sns.boxplot(data=stats_df, x = 'Type 1', y = 'Attack')
plt.show()
```

## violin plot

```python
sns.violinplot(x = 'Type 1', y = 'Attack', data=df)
```

```python
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]

sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors)
plt.show()

```

## swarm plot

```python
sns.swarmplot(x='Type 1', y='Attack', data=df, 
              palette=pkmn_type_colors)
plt.show()
```

## nakładanie wykresów

```python
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='black')# czarne punkty

sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # usunięcie boxplotów ze środka
               palette=pkmn_type_colors)
 

# Tytuł z matplotlib
plt.title('Attack by Type')
plt.show()
```

```python
melted_df = pd.melt(stats_df, 
                    id_vars=["Name", "Type 1", "Type 2"], # Zmienne do zatrzymania
                    var_name="Stat") # Nazwa nowej zmiennej
melted_df
```

```python
sns.boxplot(x='Stat', y='value', data=melted_df, 
              hue='Type 1')
plt.show()
```

```python
# powiększamy wykres
plt.figure(figsize=(16,7))
 
sns.boxplot(x='Stat', 
              y='value', 
              data=melted_df, 
              hue='Type 1', 
              palette=pkmn_type_colors) # dodajemy paletę

# przenosimy legendę
plt.legend(bbox_to_anchor=(1, 1))
plt.show()

```

## panele

```python
g = sns.catplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  # kolor
                   col='Stage',  # podział po Stage
                   kind='box') # typ
 
g.set_xticklabels(rotation=-45)
 
plt.show()
```


d) procent osób, które przeżyły (survived) w każdej z klas

```python
df = pd.crosstab(titanic["class"], titanic["survived"])
df = df.loc[:,].div(df.sum(axis=1), axis=0)
df.plot.bar()
```
