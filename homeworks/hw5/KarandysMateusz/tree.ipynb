```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def star_x(): return np.random.normal(1, 0.05, size = 500)
def star_y(): return np.random.normal(1.8, 0.05, size = 500)
def branch_x(): return np.random.normal(1, 0.2, size = 1000)
def branch_y(shift = 0): return np.random.exponential(0.07, size = 1000) + shift
def trunk_x(): return np.random.uniform(0.85, 1.15, size = 500)
def trunk_y(): return np.random.uniform(0, 0.5, size = 500)
```

```python
df = pd.DataFrame()
star = pd.DataFrame({"x": star_x(), "y": star_y(), "type": "star"})
branches = pd.DataFrame({
           "x": np.concatenate([branch_x(), branch_x(), branch_x(), branch_x(), branch_x()]),
           "y": np.concatenate([branch_y(1.4), branch_y(1.15), branch_y(0.9), branch_y(0.65), branch_y(0.4)]),
           "type": "branch"})
trunk = pd.DataFrame({"x": trunk_x(), "y": trunk_y(), "type": "trunk"})
df = pd.concat([df, trunk, branches, star])
```

```python
colors = {"star": "yellow", "branch": "green", "trunk": "brown"}
plt.scatter(df.x, df.y, c = df.type.map(colors))
plt.xlim(0, 2)
plt.ylim(0, 2)
plt.axis("off")
plt.show()
plt.savefig("tree.png")
```
